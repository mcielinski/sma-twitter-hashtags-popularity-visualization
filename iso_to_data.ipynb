```python
import pycountry
import pandas as pd
import numpy as np
```

### ADD 3-DIGIT ISO CODE TO DATA

```python
hashtags_count_df = pd.read_csv('out/count_of_COVID_hashtags_over_countries.csv')

countries_df = pd.DataFrame(hashtags_count_df['country'])
countries_df
```

```python
input_countries = countries_df['country'].tolist()

countries = {}
for country in pycountry.countries:
    countries[country.name] = country.alpha_3

codes = [countries.get(country, 'Unknown code') for country in input_countries]

codes
```

```python
hashtags_count_df['iso'] = np.array(codes)
#hashtags_count_df = hashtags_count_df.drop('Unnamed: 0', 1)
hashtags_count_df
```

```python
#SAVE COUNTS DATAFRAME TO CSV
hashtags_count_df.to_csv('out/count_of_COVID_hashtags_over_countries_iso_.csv')
```
